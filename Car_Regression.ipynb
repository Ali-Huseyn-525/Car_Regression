```python
#Covert USD and EUR to AZN

df.loc[df['Valyuta'] == 'USD', 'Qiymet'] = df.loc[df['Valyuta'] == 'USD', 'Qiymet'] * 1.7
df.loc[df['Valyuta'] == 'EUR', 'Qiymet'] = df.loc[df['Valyuta'] == 'EUR', 'Qiymet'] * 1.83
```

```python
#2.1.  Doing some split and cleaning process for 'Mühərrik'

print(df.Mühərrik.unique())
print(df.Mühərrik.nunique())
```

```python
# Split data as '/'

df[['Engine', 'Power', 'Fuel']] = df['Mühərrik'].str.split('/', expand=True)
```

```python
#Delete columns of 'Mühərrik' after splitting

df.drop(columns=['Mühərrik'], inplace=True)
```

```python
#Cleaning proces for new columns: convert -> string to int or float

df['Engine'] = df['Engine'].str.replace('L', '', regex=False).str.strip()
df['Engine'] = df['Engine'].astype(float)
```

```python
df['Power'] = df['Power'].str.replace('a.g.', '', regex=False).str.strip()
df['Power'] = df['Power'].astype(int)
```

```python
df['Yürüş'] = df['Yürüş'].str.replace('km', '', regex=False).str.strip()
df['Yürüş'] = df['Yürüş'].str.replace(' ', '', regex=False)
df['Yürüş'] = df['Yürüş'].astype(int)
```

```python
#2.2. Doing some split and cleaning process for 'Vəziyyəti'

df[['Damage', 'Color']] = df['Vəziyyəti'].str.split(',', expand=True)
```

```python
#Delete 'Vəziyyəti' columns after splitting

df.drop(columns=['Vəziyyəti'], inplace=True)
```

```python
#Change string values to string for encoding in rows

# 1.
def convert_color(x):
  x = str(x).lower()

  if 'rənglənib' in x:
    return 1

  elif 'rənglənməyib' in x:
    return 0

df['Color'] = df['Color'].apply(convert_color)
```

```python
#2.
def convert_damage(x):
  x = str(x).lower()

  if 'vuruğu yoxdur' in x:
    return 0

  elif 'vuruğu var' in x:
    return 1

df['Damage'] = df['Damage'].apply(convert_damage)
```

```python
#Check again IQR

# Select only numerical columns
numeric_df = df.select_dtypes(include='number')

# List to store column names with outliers
columns_with_outliers = []

# Check each column for outliers (with nunique > 2)
for col in numeric_df.columns:
    if numeric_df[col].nunique() > 2 and col != 'Qiymet':  # Only consider columns with more than 2 unique values
        Q1 = numeric_df[col].quantile(0.25)
        Q3 = numeric_df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        # Check if any outliers exist
        if ((numeric_df[col] < lower) | (numeric_df[col] > upper)).any():
            columns_with_outliers.append(col)

print("Columns with outliers:", columns_with_outliers)
```

```python
#Solve outliers problems with clip values methods

for col in columns_with_outliers:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Clip values outside bounds
    df[col] = np.clip(df[col], lower, upper)
```

```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

num_cols = df.select_dtypes(include='number').columns.tolist()
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

preprocessor = ColumnTransformer(transformers=[
    ('num', SimpleImputer(strategy='most_frequent'), num_cols),
    ('cat', SimpleImputer(strategy='most_frequent'), cat_cols)
])

df_processed_array = preprocessor.fit_transform(df)

final_columns = num_cols + cat_cols

df_processed = pd.DataFrame(df_processed_array, columns=final_columns, index=df.index)

df_processed[num_cols] = df_processed[num_cols].apply(pd.to_numeric)
df_processed[cat_cols] = df_processed[cat_cols].astype(str)

df_processed.head()
```

```python
cat_cols1 = df_processed.select_dtypes(include=['object', 'category']).columns

df_processed[cat_cols1].nunique()
```

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df_processed.copy()

cat_cols = df_encoded.select_dtypes(include=['object', 'category']).columns


for col in cat_cols:
    n_unique = df_encoded[col].nunique()

    if  3 <= n_unique <= 15:
        dummies = pd.get_dummies(df_encoded[col], prefix=col, drop_first=True)
        df_encoded = pd.concat([df_encoded.drop(col, axis=1), dummies], axis=1)

    elif n_unique > 15:
        freq_map = df_encoded[col].value_counts(normalize=True)
        df_encoded[col] = df_encoded[col].map(freq_map)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_encoded['Power'] = scaler.fit_transform(df_encoded['Power'].values.reshape(-1, 1))
df_encoded['Car_age'] = scaler.fit_transform(df_encoded['Car_age'].values.reshape(-1, 1))

```

```python
from sklearn.model_selection import train_test_split

X = df_encoded.drop('Qiymet', axis=1).values
y = df_encoded['Qiymet'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

poly_reg = LinearRegression()
poly_reg.fit(X_train_poly, y_train)
y_pred_poly = poly_reg.predict(X_test_poly)

mse_poly = mean_squared_error(y_test, y_pred_poly)
r2_poly = r2_score(y_test, y_pred_poly)

print("  Test MSE:", mse_poly)
print("  Test R² Score (Accuracy):", r2_poly)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
from sklearn.metrics import r2_score
from xgboost import XGBRegressor

# Initialize model
xgb = XGBRegressor()

# Learning curve
train_sizes, train_scores, val_scores = learning_curve(
    estimator=xgb,
    X=X_train,
    y=y_train,
    cv=5,
    scoring='r2',
    n_jobs=-1,
    shuffle=True,
    train_sizes=np.linspace(0.1, 1.0, 10),
    random_state=42
)

# Mean and std
train_scores_mean = np.mean(train_scores, axis=1)
val_scores_mean = np.mean(val_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
val_scores_std = np.std(val_scores, axis=1)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_scores_mean, 'o-', label='Training R²')
plt.plot(train_sizes, val_scores_mean, 's-', label='Validation R²')

plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.1)
plt.fill_between(train_sizes, val_scores_mean - val_scores_std,
                 val_scores_mean + val_scores_std, alpha=0.1)

plt.title('Learning Curve for XGBRegressor')
plt.xlabel('Training Set Size')
plt.ylabel('R² Score')
plt.legend(loc='best')
plt.grid(True)
plt.tight_layout()
plt.show()

# Optional: Final model training and evaluation
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
print("Test R² Score:", r2_score(y_test, y_pred_xgb))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import learning_curve
from sklearn.ensemble import StackingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor

# Define base and meta models
base_models = [
    ('lr', LinearRegression()),
    ('rf', RandomForestRegressor(n_estimators=100, random_state=42)),
    ('svr', SVR(kernel='rbf'))
]
meta_model = LinearRegression()

# Full stacking regressor
stacking_reg = StackingRegressor(estimators=base_models, final_estimator=meta_model, cv=5)

# Compute learning curve
train_sizes, train_scores, val_scores = learning_curve(
    estimator=stacking_reg,
    X=X_train,
    y=y_train,
    scoring='r2',
    cv=5,
    train_sizes=np.linspace(0.1, 1.0, 10),
    n_jobs=-1,
    shuffle=True,
    random_state=42
)

# Calculate mean and std
train_scores_mean = np.mean(train_scores, axis=1)
val_scores_mean = np.mean(val_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
val_scores_std = np.std(val_scores, axis=1)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_scores_mean, label="Training Score", marker='o')
plt.plot(train_sizes, val_scores_mean, label="Cross-Validation Score", marker='s')

plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.1)
plt.fill_between(train_sizes, val_scores_mean - val_scores_std,
                 val_scores_mean + val_scores_std, alpha=0.1)

plt.title("Learning Curve for Stacking Regressor")
plt.xlabel("Training Set Size")
plt.ylabel("R² Score")
plt.legend(loc="best")
plt.grid(True)
plt.tight_layout()
plt.show()



```

```python
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error


kf = KFold(n_splits=5, shuffle=True, random_state=42)
model = RandomForestRegressor(random_state=42)

mse_scores = []

for train_idx, test_idx in kf.split(X):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    mse_scores.append(mse)

print(f"Average MSE: {np.mean(mse_scores):.4f}")
```

```python
from sklearn.model_selection import cross_validate

scoring = ['neg_mean_squared_error', 'r2']
results = cross_validate(model, X, y, cv=5, scoring=scoring)

print("Average MSE:", -np.mean(results['test_neg_mean_squared_error']))
print("Average R²:", np.mean(results['test_r2']))
```
